```python
import pandas as pd
import numpy as np
import pickle
data = pd.read_csv("cleaned_data.csv", sep="\t")
```

```python
#Ha mai utilizzato il servizio di counseling psicologico offerto dall’università 
data.loc[data["D_Utilizzato"]=="si"]
```

```python
#Gruppo 1 : Voleva usare il servizio ma non l'ha fatto
G1 = data.loc[np.logical_and(data["D_Utilizzato"] == "no", data["D_Utilizzo"]=="si")]
```

```python
#Gruppo 2 : Ha usato il servizio
G2 = data.loc[(data["D_Utilizzato"] == "si")]
```

```python
anagrafica=["D_genere", "D_eta", "D_Cdl", "D_Status", "R_AreeDisciplinari"]
```

```python
#27 A che tipo di bisogni secondo lei dovrebbe rispondere un buon servizio di counseling universitario? 
domanda_27 = ["D_Tipologia_Bisogni"]
```

```python
#28 Quali tipologie di attività dovrebbe garantire un buon servizio di counseling universitario?
domanda_28 = ["D_Tipologia_Attivita"]
```

```python
#29 A quali studenti dovrebbe rivolgersi prevalentemente un buon servizio di counseling universitario?
domanda_29 = ["D_PopolazioneStudenti_1", "D_PopolazioneStudenti_2", "D_PopolazioneStudenti_3"]
```

```python
gruppo_1 = G1[anagrafica+domanda_27+domanda_28+domanda_29]
```

```python
gruppo_2 = G2[anagrafica+domanda_27+domanda_28+domanda_29]
```

```python
gruppo_1= gruppo_1.replace("mancante",np.nan)
gruppo_1.isnull().sum()
```

```python
gruppo_2= gruppo_2.replace("mancante",np.nan)
gruppo_2.isnull().sum()
```

```python
gruppo_1 = gruppo_1[gruppo_1['R_AreeDisciplinari'].notna()]
gruppo_2 = gruppo_2[gruppo_2['R_AreeDisciplinari'].notna()]
```

```python
gruppo_1.isnull().sum()
```

```python
gruppo_2.isnull().sum()
```

```python
gruppo_1['D_eta'] = gruppo_1['D_eta'].apply(pd.to_numeric)
gruppo_2['D_eta'] = gruppo_2['D_eta'].apply(pd.to_numeric)
```

```python
print(gruppo_1['D_eta'].mean())
print(gruppo_2['D_eta'].mean())
```

```python
gruppo_1['D_eta'].fillna((gruppo_1['D_eta'].mean()), inplace=True)
gruppo_2['D_eta'].fillna((gruppo_2['D_eta'].mean()), inplace=True)
```

```python
pop1 = pd.get_dummies(gruppo_2["D_PopolazioneStudenti_1"],drop_first=False)
pop2 = pd.get_dummies(gruppo_2["D_PopolazioneStudenti_2"],drop_first=False)
pop3 = pd.get_dummies(gruppo_2["D_PopolazioneStudenti_3"],drop_first=False)
pop_stud = pd.concat([pop1, pop2, pop3], axis=1)
pop_stud = pop_stud.sum(axis=1, level=0)
gruppo_2.drop(['D_PopolazioneStudenti_1', 'D_PopolazioneStudenti_2','D_PopolazioneStudenti_3'], axis=1, inplace=True)
gruppo_2 = pd.get_dummies(gruppo_2)
gruppo_2_enc = pd.concat([gruppo_2, pop_stud], axis=1)
```
